```python
# Q1. What is Abstraction in OOps? Explain with an example.
# # Abstract class is used for making skleton before coding
"""
Abstraction in OOP is the process of simplifying complex systems by focusing on essential features and hiding unnecessary details.
It allows you to create classes or interfaces that represent concepts or entities, providing a higher-level view of the system.
Abstraction promotes code modularity, maintainability, and scalability by encapsulating implementation details and 
providing simplified interfaces for interaction.
"""
import abc

class pwskills:
    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def student_assignment(self):
        pass
    @abc.abstractmethod
    def student_marks(self):
        pass
    
class studentDetails(pwskills):
    def student_details(self):
        return "This is a meth for taking student details"
    def student_assignment(self):
        return "This is a method for assigning for a particular student"
    
class data_science(pwskills):
    def student_details(self):
        return "This will return a student details from data science masters"
    def student_assignment(self):
        return "This will give you student assigned for the details of data science "
```
