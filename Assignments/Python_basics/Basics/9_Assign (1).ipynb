```python
# Q1. What is Abstraction in OOps? Explain with an example.
# # Abstract class is used for making skleton before coding
"""
Abstraction in OOP is the process of simplifying complex systems by focusing on essential features and hiding unnecessary details.
It allows you to create classes or interfaces that represent concepts or entities, providing a higher-level view of the system.
Abstraction promotes code modularity, maintainability, and scalability by encapsulating implementation details and 
providing simplified interfaces for interaction.
"""
import abc

class pwskills:
    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def student_assignment(self):
        pass
    @abc.abstractmethod
    def student_marks(self):
        pass
    
class studentDetails(pwskills):
    def student_details(self):
        return "This is a meth for taking student details"
    def student_assignment(self):
        return "This is a method for assigning for a particular student"
    
class data_science(pwskills):
    def student_details(self):
        return "This will return a student details from data science masters"
    def student_assignment(self):
        return "This will give you student assigned for the details of data science "
```

```python
# Q2) Differentiate between Abstraction and Encapsulation. Explain with an example.

"""
Abstraction:
Abstraction in OOP is the process of simplifying complex systems by focusing on essential features and hiding unnecessary details.
It involves creating classes or interfaces that represent concepts or entities, providing a higher-level view of the system. 
Abstraction allows us to work with simplified interfaces and ignore the complexities of the underlying implementation.
"""

"""
The pwskills class is an abstract base class defined using the abc module. It declares three abstract methods: student_details(),
student_assignment(), and student_marks(). These methods are defined but left empty, indicating that they must be implemented
by any class that inherits from pwskills.
Two concrete classes, studentDetails and data_science, inherit from pwskills. They provide the necessary implementations
for the abstract methods specific to their respective contexts. By inheriting from the abstract base class, these classes must 
implement the abstract methods, ensuring that the essential functionalities are available.
"""
import abc

class pwskills:
    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def student_assignment(self):
        pass
    @abc.abstractmethod
    def student_marks(self):
        pass
    
class studentDetails(pwskills):
    def student_details(self):
        return "This is a meth for taking student details"
    def student_assignment(self):
        return "This is a method for assigning for a particular student"
    
class data_science(pwskills):
    def student_details(self):
        return "This will return a student details from data science masters"
    def student_assignment(self):
        return "This will give you student assigned for the details of data science "

    
# Encapsulation:
"""
Encapsulation is the practice of bundling related data and behaviors into a single unit, typically a class. It involves
hiding internal details and exposing only necessary interfaces or methods to interact with the object. Encapsulation 
helps in achieving data integrity, modularity, and code reusability.
"""
class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if self._balance >= amount:
            self._balance -= amount
        else:
            print("Insufficient funds!")

    def get_balance(self):
        return self._balance
"""
et's consider the BankAccount class. To encapsulate the data and behaviors related to a bank account, we can define 
the attributes (account_number, balance) as private by using naming conventions (e.g., _account_number, _balance) and 
providing public methods to interact with them.
"""
```

```python
# Q3. What is abc module in python? Why is it used?
"""
The abc module in Python stands for "Abstract Base Classes". This module provides the infrastructure for defining custom abstract
base classes (ABCs).
An abstract base class is a class that cannot be instantiated, meaning you cannot create an object or instance from it. 
Its main purpose is to be a base class for other classes, providing a common API for its derivatives.

The abc module is used to create abstract base classes and to check whether a class or instance provides a particular 
interface or not. By using ABCs, developers can write classes that define a common API. If other classes implement the same API, 
they can be used interchangeably, which promotes modularity and code reuse.

The abc module provides the @abstractmethod
"""
```
