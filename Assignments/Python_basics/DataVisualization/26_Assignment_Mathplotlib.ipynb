```python
import matplotlib.pyplot as plt
"""
Q1: What is Matplotlib? Why is it used? 
Name five plots that can be plotted using the Pyplot module of Matplotlib.
"""
# Matplotlib is a Python library used for data visualization. 
# It provides a wide range of plotting capabilities through its Pyplot module.

# 1) Line plot 

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
plt.plot(x, y)
plt.show()

# 2) Bar plot:
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values)
plt.show()

# 3) Scatter plot:
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
plt.scatter(x, y)
plt.show()

# 4) Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data)
plt.show()

# 5) Pie chart:
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 15, 35]
plt.pie(sizes, labels=labels)
plt.show()

```

```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. 
Using this generated data plot a scatter plot.
"""

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
"""
The subplot() function in Matplotlib is used to create multiple plots in a single figure. 
It allows you to divide the figure into a grid of rows and columns and specify the position of 
each subplot within the grid. This is useful when you want to display multiple plots side by side 
or in a specific arrangement.
"""

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

```python
"""
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot 
and a horizontal bar plot.
"""

"""
A bar plot is a type of chart that represents categorical data with rectangular bars, where the 
length of each bar corresponds to the value or frequency of a category. It is commonly used to 
compare and visualize data across different categories.

Bar plots are useful for:
    1)Comparing values: Bar charts allow for quick visual comparison of values between different 
                        categories, making it easy to identify differences, trends, or patterns
    2)Showing rankings: Bar charts can display the ranking of categories based on their values
                        providing a clear representation of the relative magnitudes or sizes.
    3)Presenting distributions: Bar charts can represent the distribution of data across
                                categories, helping to understand the spread, concentration, or 
                                frequency of values within each category.
    4)Highlighting relationships: Bar charts can illustrate relationships between different 
                                  variables or factors, making it easier to interpret and analyze 
                                  data in a graphical format


"""

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
"""
Q5: What is a box plot? Why is it used? 
"""

"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of the 
distribution of a dataset. It displays summary statistics such as the median, quartiles, and 
outliers in a compact manner.

Box plots are used for the following purposes
1) Visualizing distribution: Box plots provide a concise summary of the data distribution, 
                            including the range, skewness, and presence of outliers.
2) Comparing datasets: Box plots are useful for comparing distributions between different groups or 
                      categories, allowing for easy identification of differences and similarities.

3) Detecting outliers: Box plots highlight outliers in the dataset, helping to identify potential 
                      anomalies or extreme values.

4) Assessing symmetry and skewness: By examining the position and length of the box and whiskers,
                                    box plots provide insights into the symmetry or skewness of 
                                    the data distribution.
"""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Values")
plt.show()
```
