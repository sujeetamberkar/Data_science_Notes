```python
"""
Que 1: Name any five plots that we can plot using the Seaborn library. 
Also, state the uses of each plot.
"""

"""
Seaborn is a powerful Python data visualization library built on top of Matplotlib. 
It provides a high-level interface for creating informative and visually appealing 
statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1) Scatter Plot:
    Uses: Displays the relationship between two continuous variables. It helps visualize patterns, 
    clusters, or correlations between variables.

2)Line Plot:
        Uses: Shows the trend or relationship between two continuous variables over a continuous 
        interval. It is commonly used to visualize time series data or to display changes over time.

3)Bar Plot:
        Uses: Compares categorical data by displaying the heights or lengths of bars representing 
        each category. It is useful for visualizing and comparing data across different categories
        or groups.

4)Histogram:
        Uses: Visualizes the distribution of a single numerical variable by dividing it into 
        bins and displaying the frequency or count of observations falling into each bin.

5)Heatmap:
        Uses: Displays a matrix-like representation of data using colors. It is often used to 
        visualize correlations between variables or to highlight patterns in a dataset.
"""
```

```python
"""
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
"""

import seaborn as sns
fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

```

```python
"""
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.
Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create box plots for 'age' and 'fare' based on 'pclass'
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')

sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')
```

```python
"""
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
"""
import seaborn as sns
import matplotlib.pyplot as plt
diamonds_data = sns.load_dataset("diamonds")

sns.histplot(data=diamonds_data, x='price', hue='cut')

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Price with Cut')
```

```python
"""
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
"""
import seaborn as sns
iris_data = sns.load_dataset("iris")
sns.pairplot(data=iris_data, hue='species')
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
import seaborn as sns
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap')
```
