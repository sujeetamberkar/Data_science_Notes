```python
"""
Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.
"""
import seaborn as sns
import plotly.express as px

titanic_data = sns.load_dataset("titanic")

scatter_plot = px.scatter(titanic_data, x="age", y="fare")
scatter_plot.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

import plotly.express as px

# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a box plot for total bill and day of the week
box_plot = px.box(tips_data, x="day", y="total_bill")

# Customize the plot
box_plot.update_traces(boxpoints='all', jitter=0.3, marker=dict(size=4, opacity=0.6))
box_plot.update_layout(
    title="Total Bill Distribution by Day of the Week",
    xaxis_title="Day",
    yaxis_title="Total Bill",
    font=dict(family="Arial", size=14),
    plot_bgcolor='rgb(255, 255, 255)'
)

# Show the box plot
box_plot.show()
```

```python
"""
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
"""
import plotly.express as px
# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a histogram for total bill, with sex on the x-axis, color-coded by day, and facetted by smoker
histogram = px.histogram(tips_data, x="sex", y="total_bill", color="day", facet_col="smoker")

# Customize the plot
histogram.update_layout(
    title="Total Bill Histogram by Sex, Colored by Day",
    xaxis_title="Sex",
    yaxis_title="Total Bill",
    font=dict(family="Arial", size=14),
    plot_bgcolor='rgb(255, 255, 255)'
)

# Show the histogram
histogram.show()
```

```python
"""
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
"""
import plotly.express as px

# Load the iris dataset from Plotly
iris_data = px.data.iris()

# Create a scatter matrix plot
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Iris Scatter Matrix Plot",
    labels={"species": "Species"},
    opacity=0.7,
    symbol="species",
    symbol_map={"setosa": "circle", "versicolor": "square", "virginica": "diamond"},
    width=700,
    height=500
)

# Update layout and axes titles
scatter_matrix.update_layout(
    font=dict(size=12),
    plot_bgcolor="white",
    title_font_size=20,
    title_font_color="black",
    legend_title_font_color="black"
)
scatter_matrix.update_traces(diagonal_visible=False)

# Show the scatter matrix plot
scatter_matrix.show()

```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.

"""
A distplot is a plot that shows the distribution of a single variable in a dataset.
It combines a histogram and a line plot to give an overview of how the data is
spread out and where the majority of values lie. It helps in understanding the shape, center, and spread of the data distribution.


there isn't a specific function called distplot like in Seaborn
"""

# We can plot displot using seaborn
import seaborn as sns

# Load the tips dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Plot a distplot for total bill
sns.distplot(tips_data["total_bill"])
```
