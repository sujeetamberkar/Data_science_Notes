```python
"""
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
"""

"""
In Bokeh, glyphs are essentially the fundamental visual components that allow us to construct various types of plots.
Glyphs can take on many forms, including circles, squares, lines, among other shapes. Each glyph is characterized by a series of
attributes like its location, dimensions, color, and level of opacity, which can be individually adjusted to alter its display.

you can employ one of the glyph methods such as circle(), line(), or rect() on the figure object.
 These methods require the data and glyph properties as arguments.
"""

import bokeh.plotting as bp
import bokeh.io as bi
from bokeh.sampledata.iris import flowers

# Define the figure object
fig = bp.figure(title='Iris Dataset Scatter Plot', x_axis_label='Length of Petal',
                y_axis_label='Width of Petal')

# Incorporate a circle glyph into the plot
fig.circle(x=flowers["petal_length"], y=flowers["petal_width"],
           size=10, color='navy', alpha=0.5)

# Render the plot
bi.show(fig)

```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
from bokeh.plotting import figure, output_file, show
from bokeh.models import Legend

# Sample data
x = [2,3,4,5,6]
y = [3,4,5,6,7]

# Output to a static HTML file
output_file("line.html")

# Create a new plot with specific title and axis labels
p = figure(title="Customized Bokeh Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Change title attributes
p.title.text_color = "darkblue"
p.title.text_font = "helvetica"
p.title.text_font_style = "bold"
p.title.align = "center"

# Add a scatter renderer with custom attributes
scatter = p.scatter(x, y, fill_color='red', size=20)

# Create a legend
legend = Legend(items=[
    ("red points", [scatter])
], location="top_right")

# Customize axes
p.xaxis.axis_label_text_color = "orange"
p.yaxis.axis_label_text_color = "green"
p.xaxis.major_label_text_color = "blue"
p.yaxis.major_label_text_color = "purple"

# Add legend to the plot
p.add_layout(legend)

# Display the plot
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
"""
The Bokeh server, part of the Bokeh visualization library in Python, enables you to produce interactive plots and applications.
It empowers developers to create web-based, real-time updated data visualizations.
This means that the server can drive web applications that dynamically react to user actions
like zooming, panning, and data filtering, without needing a page refresh.

To utilize the Bokeh server for an interactive plot, you must first establish your plot within a Bokeh Document object.
This object will hold all the Bokeh components constructing your plot, including figures, glyphs, and layouts.
After setting up your plot, you can integrate interactive widgets, such as sliders, buttons, and dropdown menus, using Bokeh's models API.

Once your plot and widgets are set, you can kick off the Bokeh server with the bokeh serve command.
This action generates a web application that hosts your interactive plot and widgets.
As users interact with these widgets, the server will instantaneously refresh the plot based on user input.
"""

from bokeh.io import curdoc
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Slider
from bokeh.plotting import figure

# Define the data source
source = ColumnDataSource(data=dict(x=[1, 2, 3], y=[1, 4, 9]))

# Create the plot
plot = figure(title='Interactive Plot', x_axis_label='X', y_axis_label='Y')
plot.line('x', 'y', source=source, line_width=2)

# Set up the slider widget
slider = Slider(title='Power', value=1, start=0.1, end=5, step=0.1)

# Create the callback function
def update_data(attrname, old, new):
    new_y = [x**slider.value for x in source.data['x']]
    source.data = dict(x=source.data['x'], y=new_y)

slider.on_change('value', update_data)

# Combine the plot and slider in a layout
layout = column(plot, slider)

# Add the layout to the current document
curdoc().add_root(layout)

```

```python
# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

"""
1) Start by importing the required modules from Flask and Bokeh in your Python script.
    This includes Flask, render_template from Flask, and components, figure, and CDN from Bokeh.
2) Set up your Flask app instance. After that, define a route ('/') where the Bokeh plot will be served.
3) Within this route, create your Bokeh plot using the figure function and add any graphical elements you need. In this case, a line graph is added.

4) Subsequently, use the components function from Bokeh to generate the HTML components needed to embed the plot into a web page.
    These components include a script that enables interactivity and a div that holds the plot.

5) Finally, use Flask's render_template function to render the web page. Pass in the name of the HTML template and the script, div, and CDN links as arguments
"""
from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure
from bokeh.resources import CDN

# Initialize Flask app
app_instance = Flask(__name__)

@app_instance.route('/')
def home():
    # Construct the Bokeh plot
    plot_object = figure(title='My Plot', x_axis_label='X', y_axis_label='Y')
    plot_object.line([1, 2, 3], [1, 4, 9])

    # Create the HTML components
    bokeh_script, bokeh_div = components(plot_object)

    # Use the components in the rendered template
    return render_template('index.html', script=bokeh_script, div=bokeh_div, cdn_js=CDN.js_files[0], cdn_css=CDN.css_files[0])

if __name__ == '__main__':
    app_instance.run()

```
