```python
"""
We want to check the relation between steple length and steple width 

Is steple length directly propotional to steple width ???
"""
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
# This shows that there is no relationship 
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
# There are few flowers where the petal lenth is sirectly proportional to sepal negth 
# But there are also some flowers where the seplanth is not directly proportional to petal length  @@@ Constant 
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# Lets check if there is any relation between tip and the bil 
# The person who spends more pays more tips ???
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.head()
```

```python
# Use normal pandas to check smokers and non non smokers
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips.total_bill, y = tips.tip, data=tips,hue='smoker')
# colour coding according to category 
# In hue you put the name of any column 
# In smoker, we have 2 column 
# Yes and no 
```

```python
tips.head()
```

```python
# Check if the person pays more bil;l when tehy come in group of 2,3,4.. etc 
sns.relplot(x=tips.total_bill, y = tips.tip, data=tips,hue='size')
```

```python
tips['size'].value_counts()
```

```python
# If you want different style instead of different colour 
sns.relplot(x=tips.total_bill, y = tips.tip, data=tips,style='size')
```

```python
# If you want different style instead of different colour 
sns.relplot(x=tips.total_bill, y = tips.tip, data=tips,hue='time',style='size')
```

```python
# Which day more staff is needed ?
# WHich day is more crowdded ?
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
# Highest bill was on saturday 
# Most number of people (Highest density ) is on saturday 
# Friday very few people 
```

```python
# WHo gives more tips ???
# Smoker or non smoker 

sns.catplot(x='smoker',y='total_bill',data=tips)
```

```python
sns.jointplot(x = tips.total_bill , y = tips.tip)
# So usually average people tip 2-4 dolors 
```
