```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Make data for x axis and y axis respectively 
# Then plot the dat a
x= np.linspace(0,10,200)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Generate random data and plot it 
# Only dots 
x= np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y)
```

```python
# Make sure that all dots are of different colour 
# Of different sizes and colours
sizes=100*np.random.rand(50)
colours=np.random.rand(50)
x= np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes)
```

```python
# Intensity of colour is controlled by alpha 
sizes=100*np.random.rand(50)
colours=np.random.rand(50)
x= np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.3)
```

```python
sizes=100*np.random.rand(50)
colours=np.random.rand(50)
x= np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.3)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Bar graph 
x=['a','b','c','d','e']
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Horizontal Bar graph 
x=['a','b','c','d','e']
y=np.random.rand(5)
plt.barh(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Lets play with sine graph 
x= np.linspace(0,10,200)
y=np.sin(x)
plt.plot(x,y,'--')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Control size
# Lets play with sine graph 
x= np.linspace(0,10,200)
y=np.sin(x)
plt.figure(figsize=(2,4))
plt.plot(x,y,'--')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.show()
```

```python
# Grid 
x= np.linspace(0,10,200)
y=np.sin(x)
plt.figure(figsize=(2,4))
plt.plot(x,y,'--')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data ")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
# 3d data 

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure() # Make a figure
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
